```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", device='cuda')

input_text = "Astronomy is the study of stars and planets."
possible_labels = ["Science", "History", "Sports"]

result = classifier(input_text, possible_labels)

print(result)

print("Input Text:", input_text)
print("Predicted Class:", result["labels"][0])
print("Confidence Score:", result["scores"][0])
```

```python
import json

def get_reviews(worker_id):
    dataset_path = r"M:\python_projects\innoglobalhack\InnoglobalHack\dataset\review_dataset.json"

    with open(dataset_path, 'r', encoding='utf-8') as file:
        ds_reviews = json.load(file)

    return [
        item['review']
        for item in ds_reviews
        if item['ID_under_review'] == worker_id
        and item['ID_reviewer'] != worker_id
    ]

rev = get_reviews(6135)
rev
```

```python
criteria_labels = {
    'usefulness': ['полезный', 'не полезный'],
    'content': ['содержательный', 'не содержательный'],
    'objectivity': ['объективный', 'не объективный'],
    'ethic': ['этичный', 'не этичный'],
    'honesty': ['честный', 'не честный'],
    'detail': ['подробный', 'не подробный']
}
```

```python
def evaluate_review(review_text):
    total_score = 0.0
    scores = {}
    
    for criterion, labels in criteria_labels.items():
        result = classifier(review_text, labels)
        print(result)
        
        positive_label_index = result['labels'].index(labels[0])
        
        positive_score = result['scores'][positive_label_index]
        if criterion == "detail":
            positive_score /= 3
        
        scores[criterion] = positive_score
        total_score += positive_score
    
    return {'total_score': total_score, 'scores': scores}
```

```python
evaluate_review(rev[0])
```

```python
evaluate_review("Хороший специалист")
```

```python
review = "Отличный профессионал, подключается к решению общих задач, в том числе и по смежным темам. Всегда готов договариваться и находить хорошее решение для всех заинтересованных сторон. Это очень ценный подход со стороны функции информационной безопасности."
evaluate_review(review)
```

```python
evaluated_reviews = []
    
for review in rev:
    result = evaluate_review(review)
    result['review'] = review
    evaluated_reviews.append(result)
    
sorted_reviews = sorted(evaluated_reviews, key=lambda x: x['total_score'], reverse=True)

for review in sorted_reviews:
    print(review['total_score'], review['review'])
```
