```python
from transformers import BartForConditionalGeneration, BartTokenizer
import time

# Загрузка модели и токенизатора
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')

# Текст для суммаризации
text = """
В последние годы наблюдается рост интереса к искусственному интеллекту и его применениям в различных сферах жизни. 
От медицины до финансов, AI помогает улучшить процессы и повысить эффективность. 
Однако с этим ростом также возникают вопросы этики и безопасности, которые необходимо учитывать.
"""

s = time.time()

# Токенизация текста
inputs = tokenizer(text, return_tensors='pt', max_length=1024, truncation=True)

# Генерация суммари
summary_ids = model.generate(inputs['input_ids'], max_length=150, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)

# Декодирование и вывод результата
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

e = time.time()

print(summary)
print(e-s)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Load the tokenizer and model
model_name = "sn4kebyt3/ru-bart-large"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Initialize a summarization pipeline
summarizer = pipeline("summarization", model=model, tokenizer=tokenizer)

# Input text in Russian
text = """
Ваш текст на русском языке здесь. Например, это может быть длинный абзац из статьи, и вы хотите получить краткое изложение этой информации.
"""

# Generate summary
summary = summarizer(text, max_length=150, min_length=50, do_sample=False)

# Output the summarized text
print("Summary:", summary[0]['summary_text'])
```
